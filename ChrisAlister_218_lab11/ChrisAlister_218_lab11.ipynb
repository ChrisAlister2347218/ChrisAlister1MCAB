Perform the Exploratory Data Analysis on your domain-based dataset and demonstrate
the retrieved insights using “Matplotlib” modules. Visualize hidden insights using appropriate
plots (graphs) [Usage of line plot and scatter plot are mandatory]

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('ChrisAlister_218_Dataset.csv')

# Display the first few rows of your dataset to understand its structure
print(df.head())

# Summary statistics
print(df.describe())

# Scatter plot (Price vs. NumImages)
plt.figure(figsize=(8, 6))
plt.scatter(df['Price (INR)'], df['NumImages'], color='blue', label='Price vs. NumImages')
plt.title('Scatter Plot of Price vs. NumImages')
plt.xlabel('Price (INR)')
plt.ylabel('NumImages')
plt.legend()
plt.show()

# Line plot (Gender vs. Price)
gender_counts = df['Gender'].value_counts()
gender_labels = gender_counts.index
gender_values = gender_counts.values

plt.figure(figsize=(8, 6))
plt.plot(gender_labels, gender_values, marker='o', linestyle='-', color='green')
plt.title('Line Plot of Gender vs. Count')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.grid(True)
plt.show()

# For categorical columns like ProductName or ProductBrand, you can create bar plots:
product_brand_counts = df['ProductBrand'].value_counts()
plt.figure(figsize=(12, 6))
product_brand_counts.plot(kind='bar', color='orange')
plt.title('Bar Plot of ProductBrand vs. Count')
plt.xlabel('ProductBrand')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
